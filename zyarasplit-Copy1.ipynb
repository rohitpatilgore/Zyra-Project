```python
# Path to the large CSV file
input_file = 'df.csv'
# Number of rows per smaller CSV file
rows_per_file = 100_000
# Output file prefix
output_file_prefix = 'split_file_'

# Read the large CSV file in chunks
chunk_iter = pd.read_csv(input_file, chunksize=rows_per_file)

# Counter for file naming
file_count = 1

for chunk in chunk_iter:
    if file_count > 10:  # Stop after 10 files
        break
    # Define the output file name
    output_file = f"{output_file_prefix}{file_count}.csv"
    # Write the chunk to a new CSV file
    chunk.to_csv(output_file, index=False)
    # Increment the file counter
    file_count += 1

print("Splitting completed after 10 files.")
```

```python
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
import glob
import os
from rapidfuzz import process

# Define the path to your CSV files (update this to your actual path)
path = r'C:\Users\Rohit Subhash Gore\Documents\ZyraTech 3'
csv_files = glob.glob(os.path.join(path, "*.csv"))

# Define the names or patterns of the files to skip
files_to_skip = ['9M+ companies.csv', 'df.csv']

# Filter out the specified files from the list of CSV files
csv_files = [file for file in csv_files if not any(skip_file in file for skip_file in files_to_skip)]

# Shared list to hold the DataFrames read by each thread
dataframes = []

# Function to read a CSV file into a pandas DataFrame
def read_csv(file):
    print(f"Thread starting to read {file}")
    try:
        df = pd.read_csv(file)
        print(f"Successfully read {file}")
        return df
    except Exception as e:
        print(f"Error reading {file}: {e}")
        return None

# Number of threads to use for reading files
max_threads = min(10, len(csv_files))

# Using ThreadPoolExecutor for multithreading to read CSV files
with ThreadPoolExecutor(max_workers=max_threads) as executor:
    results = executor.map(read_csv, csv_files)

# Filter out any failed reads
dataframes = [df for df in results if df is not None]

# Concatenate the DataFrames
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    print("All files successfully combined.")
else:
    print("No DataFrames were created. Check the CSV files.")

# Function to match user input with fuzzy matching using RapidFuzz
def fuzzy_match(input_value, column_values):
    # Extract the best match and score
    result = process.extractOne(input_value, column_values)
    if result is None:
        return None
    match, score = result[0], result[1]  # Unpack only the match and score
    return match if score > 70 else None  # Use a threshold of 70% similarity

# Function to find companies using fuzzy matching
def find_companies_with_fuzzy_matching(df, industry_req, location_req, size_range_req):
    industries = df['industry'].str.lower().unique()
    localities = df['locality'].str.lower().unique()
    size_ranges = df['size_range'].str.lower().unique()

    industry_req_corrected = fuzzy_match(industry_req.lower(), industries)
    location_req_corrected = fuzzy_match(location_req.lower(), localities)
    size_range_req_corrected = fuzzy_match(size_range_req.lower(), size_ranges)

    if industry_req_corrected and location_req_corrected and size_range_req_corrected:
        matching_companies = df[
            (df['industry'].str.lower() == industry_req_corrected) &
            (df['locality'].str.lower() == location_req_corrected) &
            (df['size_range'].str.lower() == size_range_req_corrected)
        ]
        return matching_companies
    else:
        print("No close matches found for the input.")
        return pd.DataFrame()  # Return empty DataFrame if no match

# Function to handle user input and find companies based on fuzzy matching
def user_input_and_filter():
    industry_req = input("Enter the industry you're searching for: ")
    location_req = input("Enter the location you're searching for: ")
    size_range_req = input("Enter the size range you're searching for (e.g., '10001+', '5001 - 10000'): ")

    matching_companies = find_companies_with_fuzzy_matching(combined_df, industry_req, location_req, size_range_req)

    if not matching_companies.empty:
        output_csv_file = "matching_companies.csv"
        output_json_file = "matching_companies.json"

        # Save output to CSV and JSON files
        matching_companies.to_csv(output_csv_file, index=False)
        matching_companies.to_json(output_json_file, orient='records')  # Save standard JSON format

        print(f"\nCompanies matching your criteria have been saved to {output_csv_file} and {output_json_file}.")
    else:
        print("\nNo companies found matching your criteria.")

# Running the user input and filtering in a separate thread
with ThreadPoolExecutor(max_workers=1) as executor:
    executor.submit(user_input_and_filter)
```
